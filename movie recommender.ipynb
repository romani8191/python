```python
import numpy as np
import pandas as pd
movie=pd.read_csv("movies.csv")
movie
```

```python
import re
def clear_title(title):
    return re.sub("[^a-zA-Z0-9]"," ",title)
```

```python
movie["clear_title"]=movie["title"].apply(clear_title)
```

```python
movie
```

```python
from sklearn. feature_extraction. text import TfidfVectorizer
vectorizer = TfidfVectorizer (ngram_range=(1,2))
tfidf = vectorizer.fit_transform (movie ["clear_title"])
movie
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
def search(title):
    title= clear_title(title)
    query_vec= vectorizer.transform( [title])
    similarity = cosine_similarity (query_vec, tfidf).flatten()
    indices= np.argpartition (similarity, -5) [-5:]
    results = movie.iloc [indices] [::-1]
    return results
```

```python
import ipywidgets as widgets
from IPython.display import display
movie_input = widgets.Text(
    value="Toy Story",
    description="Movie Title:",
    disabled=False
)
movie_list = widgets.Output()
def on_type(data):
    with movie_list:
        movie_list.clear_output()
        title= data["new"]
        if len(title) > 5:
            display (search(title))
    
movie_input.observe (on_type, names='value')
display (movie_input, movie_list)
```

```python
def find_similar_movie(movie_id):
    #similar user recommendations
    similar_user=ratings[(ratings["movieId"]==movie_id)& (ratings["rating"]>=5)]["userId"].unique()
    similar_user_rec=ratings[  (ratings["userId"].isin(similar_user))  &  (ratings["rating"]>=4)  ] ["movieId"]
    
    #filtering for top 10% recommendations
    similar_user_rec=similar_user_rec.value_counts()/len(similar_user)
    similar_user_rec=similar_user_rec[similar_user_rec>.10]

    #looking out for other users
    all_users=ratings[(ratings["movieId"].isin(similar_user_rec.index)) & (ratings["rating"]>=4)]
    all_users_rec=all_users["movieId"].value_counts()/len(all_users["userId"].unique())

    #combining the above 2 
    rec_percentages=pd.concat([similar_user_rec,all_users_rec],axis=1)
    rec_percentages.columns=["similar","all"]
    
    #calculating n arranging the combined result
    rec_percentages["score"]=rec_percentages["similar"] / rec_percentages["all"]
    rec_percentages=rec_percentages.sort_values("score",ascending=False)
    return rec_percentages.head(10).merge(movie, left_index=True, right_on="movieId")[["score", "title", "genres"]]
```

```python
#creating the interactive widget

movie_name_input=widgets.Text(
    value="Toy story",
    description="Enter title: ",
    disabled=False
)

recommendation_list=widgets.Output()

def on_type(data):
    with recommendation_list:
        recommendation_list.clear_output()
        title=data["new"]
        if len(title)>5:
            results=search(title)
            movie_id=results.iloc[0]["movieId"]
            display(find_similar_movie(movie_id))
            
movie_name_input.observe(on_type,names="value")
display(movie_name_input,recommendation_list)
```
